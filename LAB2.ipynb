```python
import sys
import math
def serie_1(N):
    suma=0
    lista=[]
    for i in range(0,N):
        elemento=(1/(2*i+1))*((-1)**i)
        suma=suma+elemento
        lista.append(suma)
    #print("La suma total(1) es ", suma, "numero de iteraciones", i)
    return lista, suma

def serie_2(referencia, precision):
    suma=0
    i=0
    lista=[]
    while(True):
        elemento=(1/(2*i+1))*((-1)**i)
        suma=suma+elemento
        lista.append(suma)
        if(abs(suma-referencia)<precision):
            break
        i=i+1
    return lista, suma
    #print("La suma total(2) es ", suma, "numero de iteraciones", i)

def serie_3(precision):
    suma=0
    i=0
    lista=[]
    while(True):
        elemento=(1/(2*i+1))*((-1)**i)
        copia_suma=suma
        suma=suma+elemento
        lista.append(suma)
        if(i!=0):
            error=abs((suma-copia_suma)/copia_suma)
        if(i!=0 and error<=precision):
            break
        i=i+1
    return lista, suma
    # print("La suma total(3) es ", suma, "numero de iteraciones", i)

#Pregunta_1 Python 
N=4000
lista_1, suma_1= serie_1(N)
print(lista_1)
print("La suma de la serie es ", suma_1)
#Pregunta_2 Python
referencia= (math.pi)/4
precision_1=0.0001
lista_2, suma_2= serie_2(referencia, precision_1)
print(lista_2)
print("La suma de la serie es", suma_2)
#Pregunta_3 Python
precision_2= 0.00001
lista_3, suma_3= serie_3(precision_2)
print(lista_3)
print("La suma de la serie es", suma_3)
```
